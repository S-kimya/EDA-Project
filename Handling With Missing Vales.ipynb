```python
def analyze_null_values(df):
    """
    Compute null value counts and percentages for each column in a DataFrame.

    Parameters:
    df (pandas.DataFrame): Input DataFrame to analyze

    Returns:
    pandas.DataFrame: DataFrame with null count and percentage for each column
    """
    #Calculate null counts
    null_counts = df.isnull().sum()
    #Calculate null percentages
    null_percentages = (df.isnull().sum() / len(df)) * 100
    #Combine into a single DataFrame
    null_analysis = pd.DataFrame({
        'Null Count': null_counts,
        'Null Percentage': null_percentages.round(2)
    })
    return null_analysis.sort_values(by = 'Null Percentage')

null_summary = analyze_null_values(df)
null_summary
```

```python
def fill_na_with_mode(df, column, group_by_hierarchy):
    """
    Fill NA values in a specified column of a DataFrame by using the mode 
    from a hierarchical grouping approach.
    
    Parameters:
    - df (DataFrame): The input DataFrame.
    - column (str): The name of the column to fill NA values for.
    - group_by_hierarchy (list of str): A list of column names representing 
    the hierarchy for grouping (from most specific to least specific).
    """
    # Create a copy to avoid modifying the original DataFrame
    df_filled = df.copy()

    # Iterate over the group by hierarchy
    for group_by in group_by_hierarchy:
        # Filter rows that still have NA in the target column after previous iterations
        rows_to_fill = df_filled[column].isna()
        
        if rows_to_fill.any():
            # Calculate the mode for each group using all non-NA values
            mode_values = df_filled.groupby(group_by)[column].transform(
                lambda x: x.mode().iloc[0] if not x.mode().empty else None
            )
            # Fill NA values in the original DataFrame with the calculated modes
            df_filled.loc[rows_to_fill, column] = mode_values[rows_to_fill]
    
    return df_filled

```

```python
def fill_na_with_mean(df, column, group_by_hierarchy):
    """
    Fill NA values in a specified column of a DataFrame by using the mean 
    from a hierarchical grouping approach.
    
    Parameters:
    - df (DataFrame): The input DataFrame.
    - column (str): The name of the column to fill NA values for.
    - group_by_hierarchy (list of str): A list of column names representing 
    the hierarchy for grouping (from most specific to least specific).
    """
    # Create a copy to avoid modifying the original DataFrame
    df_filled = df.copy()

    # Iterate over the group by hierarchy
    for group_by in group_by_hierarchy:
        # Filter rows that still have NA in the target column after previous iterations
        rows_to_fill = df_filled[column].isna()
        
        if rows_to_fill.any():
            # Calculate the mode for each group using all non-NA values
            mean_values = df_filled.groupby(group_by)[column].transform(
                lambda x: round(x.mean()) if not x.isnull().all() else np.nan
            )
            # Fill NA values in the original DataFrame with the calculated modes
            df_filled.loc[rows_to_fill, column] = mean_values[rows_to_fill].astype(int)
    
    return df_filled
```

```python
import pandas as pd
from collections import Counter

# Fonksiyonun düzeltilmiş hali
def find_feature_mode(series):
    processed_items = []
    
    for item in series:
        if isinstance(item, str):  # Eğer öğe bir string ise
            processed_items.extend([i for i in item.split(', ') if i != "NaN"])
        elif isinstance(item, (list, tuple)):  # Eğer öğe bir liste ya da tuple ise
            processed_items.extend([i for i in item if i != 'NaN'])
        elif pd.notna(item):  # Eğer öğe NaN değilse
            processed_items.append(str(item))
    
    if processed_items:
        mode_items = Counter(processed_items).most_common(5)  # En yaygın 5 öğe
        return mode_items
    return None

# İlgili sütunlar
columns_of_interest = ['comfort_convenience', 'entertainment_media', 'extras', 'safety_security']

# DataFrame üzerinde gruplama ve işlem yapma
for name, group in df.groupby('make_model'):
    print(f"--- **Make & Model:** {name} ---")
    
    for column in columns_of_interest:
        if column in group.columns:
            mode_result = find_feature_mode(group[column])
            if mode_result:
                print(f"{column} Mode: **{mode_result[:3]}** (Count: {mode_result[:3]})")  # İlk 3 sonucu yazdır
            else:
                print(f"{column} Mode: **No Mode Found**")
    print()  # Grup arasındaki boş satır
```

```python
def count_items (series):
    """
    Counts the number of ltens In each entry of a pandas Series.
    Returns None for NaN or empty entries.
    """
    def count_logic(item):
        if isinstance(item, str):
            return len(item.split(', '))
        return 0

    return series.apply(count_logic)
df_cleaned['comfort_convenience_counts'] = count_items(df_cleaned['comfort_convenience'])
df_cleaned['entertainment_media_counts'] = count_items(df_cleaned['entertainment_media'])
df_cleaned['extra_counts'] = count_items(df_cleaned['extras'])
df_cleaned['safety_security_counts'] = count_items(df_cleaned['safety_security'])
# Display the Datafrane to check results
df_cleaned[['comfort_convenience_counts', 'entertainment_media_counts',
'extra_counts','safety_security_counts' ]]
```
